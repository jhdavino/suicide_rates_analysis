```python
import folium
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
df = pd.read_csv("master.csv")
print(df.columns)
df.sample(1)
```

```python
country_freq = df.country.value_counts()
country_freq.head()
```

```python
countries = country_freq.index.tolist()
countries[:5]
```

```python
freq = country_freq.values.tolist()
freq[:5]
```

```python
lat = []
lon = []
geolocator = Nominatim(user_agent="show_countries",timeout=10)
for c in countries:
    location = geolocator.geocode(c)
    lat.append(location.latitude)
    lon.append(location.longitude)
    
print(lat[0])
print(lon[0])
print(countries[0])
```

```python
# Make an empty map
m = folium.Map(location=[20, 0], tiles="Mapbox Bright", zoom_start=2)
```

```python
# I can add marker one by one on the map
for i in range(0,len(countries)):
    c = countries[i]
    folium.Marker([lon[i], lat[i]], popup=c).add_to(m)
```

```python
# Save it as html
m.save('312_markers_on_folium_map1.html')
```
